```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../DATASET/Churn_Modelling.csv")
df.head()
```

```python
df = pd.get_dummies(df, columns=['Geography'], drop_first=True)
df.head()
```

```python
df['Gender'] = df['Gender'].map({'Male': 1, 'Female': 0})
df.head()
```

```python
df = df.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_cols = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']

df[scaled_cols] = scaler.fit_transform(df[scaled_cols])
```

```python
df['BalanceSalaryRatio'] = df['Balance'] / (df['EstimatedSalary'] + 1)
df['AgeTenureRatio'] = df['Age'] / (df['Tenure'] + 1)
```

```python
df.head()
```

```python
# Fill Gender NaNs with 0 or mode
# If you want to assign 0 to all NaNs:
df['Gender'] = df['Gender'].fillna(0)

# OR, if Gender originally had 'Male' and 'Female', encode as 0/1
# df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})
# df['Gender'] = df['Gender'].fillna(0)  # Fill missing with 0

# Convert boolean columns to 0 and 1
df['Geography_Germany'] = df['Geography_Germany'].astype(int)
df['Geography_Spain'] = df['Geography_Spain'].astype(int)

# Check result
df.head()
```

```python
# Map Gender to integer: Female -> 0, Male -> 1
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})

# Fill NaN values with 0 (defaulting to Female)
df['Gender'] = df['Gender'].fillna(0).astype(int)
```

```python
# Save the cleaned data after feature engineering
df.to_csv("../DATASET/processed_churn_data.csv", index=False)
```
